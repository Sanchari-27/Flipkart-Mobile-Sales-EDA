### 1. **Introduction**

Flipkart Mobile sales analysis for exploratory data analysis (EDA) involves collection of mobile phone information from Flipkart. It includes details like brand, model, base color, processor type, screen size, storage (ROM), RAM, display size, camera specifications (number of rear and front cameras), battery capacity, and user ratings. It also contains financial information such as the sales price, discount percentage, number of ratings, and total sales figures for each mobile phone.

### 7.1. What is the distribution of brands in the dataset?

### 7.2. How many models are there for each and every brand?

```python
#Checking different models for each brand
brand_model_count= data['brand'].value_counts().reset_index().rename(columns={'count':'model_count'})
brand_model_count
```

**Observations**:

- From **count plot diagram** we conclude that **Poco and Apple** has the minimum number of **Model count** .

### 7.3. Which mobile models have the highest and lowest ratings?

```python
max_rating= data.groupby(['brand']).max().sort_values('ratings',ascending=False).head()
max_rating=max_rating[['model','ratings']]
max_rating
```

```python
min_rating= data.groupby(['brand']).min().sort_values('ratings',ascending=True).head()
min_rating=min_rating[['model','ratings']]
min_rating
```

**Observations**:

- The mobile model **Galaxy A03s** of the brand **Samsung** has the **lowest rating**.
- The mobile model **iPhone XR** of the brand **Apple** has the **highest rating**.

### 7.12. How do the different combinations of RAM and ROM impact sales and customer ratings?

```python
ram_rom_group = data.groupby(['RAM', 'ROM']).agg(
    avg_sales=('sales_price', 'mean'),
    avg_rating=('ratings', 'mean')
).reset_index()

# Display the grouped data
print(ram_rom_group)

# Visualization: Sales by RAM and ROM combinations
plt.figure(figsize=(12, 6))
sns.barplot(data=ram_rom_group, x='RAM', y='avg_sales', hue='ROM', palette='Set2')

plt.title('Impact of RAM and ROM on Average Sales')
plt.xlabel('RAM (GB)')
plt.ylabel('Average Sales')
plt.legend(title='ROM (GB)')
plt.show()

# Visualization: Customer Ratings by RAM and ROM combinations
plt.figure(figsize=(12, 6))
sns.barplot(data=ram_rom_group, x='RAM', y='avg_rating', hue='ROM', palette='coolwarm')

plt.title('Impact of RAM and ROM on Average Ratings')
plt.xlabel('RAM (GB)')
plt.ylabel('Average Rating')
plt.legend(title='ROM (GB)')
plt.show()
```

**Observations**:

- From **Impact of RAM and ROM on Average Sales** we can analyze that the mobile models having **12GB RAM** and **512GB ROM** has the **highest average sales**.
- And the mobile models having **2GB RAM 16GB ROM**, **2GB RAM 32GB ROM** and **3GB RAM 32GB ROM** has the **lowest average sales**.
- From **Impact of RAM and ROM on Average Ratings** we can analyze that the mobile models having **3GB RAM 100GB ROM** and **3GB RAM 200GBROM** has the **highest average rating**.
- And the mobile models having **1GB RAM 100GB ROM** has the **lowest average rating**.
